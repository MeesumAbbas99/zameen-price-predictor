```python
def PriceNormalization(price):
    if "Lakh"  in price:
        price = price.replace("Lakh",'')
        price = float(price.strip())
        return price/100
    else:
        price = price.replace("Crore",'')
        return float(price.strip())
```

```python
df['Price_processed'] = df['Price_processed'].apply(PriceNormalization)
```
